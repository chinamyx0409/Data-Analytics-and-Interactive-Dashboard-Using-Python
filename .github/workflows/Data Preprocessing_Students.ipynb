```python
#Import the required libraries

import pandas as pd
import numpy as np
```

```python
#Load the data from the link/url

df = pd.read_csv("https://bit.ly/2Qk25Xx") #loading the data into a dataframe

df.head()

```

```python
#Set Arrest key as the index

df = df.set_index('ARREST_KEY')

df.head()
```

```python
#see the list of columns

df.columns
```

```python
#drop two columns: LAW_CODE, ARRESR_BORO

columns_to_drop = ['LAW_CODE', 'ARREST_BORO']
df = df.drop(columns_to_drop, axis='columns')

df.head()
```

```python
#I need to check how many rows and columns are there in my data
#Shape

df.shape
```

```python
#How many null values are there in each column

df.isna().sum()
```

```python
df2 = df.dropna(thresh = 95500, axis = 'columns' )

df2.isna().sum()
```

```python
#Create a new column in the dataframe which is called Jail time

#If Felony (F) - 36 months
#If Misdemeanours (M) - 12 months
#If Violation (V) - 1 month
```

```python
#np.where

df['JAIL_TIME'] = np.where(df['LAW_CAT_CD']=="F", 36,
                          np.where(df['LAW_CAT_CD']=="M", 12,
                                  np.where(df['LAW_CAT_CD']=="V",1,0)))

df.head()
```
