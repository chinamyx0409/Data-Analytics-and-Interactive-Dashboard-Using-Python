```python
# importing the reuqired libaries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

```

```python
census_data = pd.read_csv("http://bit.ly/census-data-wkshp")

# show 5 rows
census_data.head()

# show 10 rows
census_data.head(10) 

# check describe
census_data.describe()

# check dimension
census_data.shape

# check data type
type(census_data)

# how to check total null values
census_data.info()

#df.sum(na)


# in order to find a relationship between the numeric cols and data, we can use heatmap
# correlation and coefficient ranges from -1 to +1, 
# values closer to -1 means closer to -1,
# values closer to 0 means neutral,
# values closer to +1 means closer to +1 (postive corellationship),

# gives the correlation of all the numeric columns
correl = census_data.corr()


# plot the correl in a heatmp
sns.heatmap(correl, annot = True)
```

```python
# There is no relationship we can identify from the heatmap
# values of the correlation coeff closer to -1/+1 means strong correlation
# values of the correaltion coef closer to 0 means weak correlationshp
```
